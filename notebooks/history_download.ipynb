```python
def download_picks(match_id, api_key):
    url = 'https://api.steampowered.com/IDOTA2Match_570/GetMatchDetails/V1/'
    data = {
      "match_id": match_id,
      "key": api_key
    }
    response = requests.get(url, params=data)
    result = response.json()['result']
    if result['game_mode'] != 2:
        return None
    picks_bans = pd.DataFrame(result['picks_bans']).set_index('order').sort_index()
    
    return picks_bans
```

```python
def download_tournament_game_history(tournament_id, api_key):
    matches = []
    matches_part = []
    last_lowest_match_id = None
    url = 'https://api.steampowered.com/IDOTA2Match_570/GetMatchHistory/v1/'

    while True:
        data = {
          "key": steam_api_key,
          "league_id" : tournament_id,
          "matches_requested": 100
        }

        if last_lowest_match_id is not None:
            data['start_at_match_id'] = last_lowest_match_id

        response = requests.get(url, params=data)
        result = response.json()['result']

        if len(result['matches']) == 0:
            break

        matches_part = pd.DataFrame(result['matches']).set_index('match_id')
        if last_lowest_match_id == min(matches_part.index):
            break

        last_lowest_match_id = min(matches_part.index)

        print('last id = {}'.format(last_lowest_match_id))
        matches.append(matches_part)
    
    res = pd.concat(matches).sort_index()
    
    return res
```

```python
def process_game(game_id):
    res = download_picks(game_id, steam_api_key)
    if res is None:
        return (game_id, None)
    match_data = [res.loc[0]['team']] +res['hero_id'].values.tolist()
#     matches_data[game_id] = match_data
    return (game_id, match_data)
```
